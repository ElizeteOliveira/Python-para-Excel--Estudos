```python
import pandas as pd
```

```python
# Lista de arquivos Excel a serem processados
arquivos_entrada = [
    r'C:\Users\E1180062\Downloads\Análise de Orçamento\Dados_Combinados_JAN_FLO.xlsx',
    r'C:\Users\E1180062\Downloads\Análise de Orçamento\Dados_Combinados_FEV_FLO.xlsx',
    r'C:\Users\E1180062\Downloads\Análise de Orçamento\Dados_Combinados_MAR_FLO.xlsx',
    r'C:\Users\E1180062\Downloads\Análise de Orçamento\Dados_Combinados_ABR_FLO.xlsx',
    r'C:\Users\E1180062\Downloads\Análise de Orçamento\Dados_Combinados_MAI_FLO.xlsx',
    r'C:\Users\E1180062\Downloads\Análise de Orçamento\Dados_Combinados_JUN_FLO.xlsx'
]
```

```python
def processar_arquivo(arquivo_entrada):
    try:
        # Leitura do arquivo Excel
        df = pd.read_excel(arquivo_entrada)
        print(f"Arquivo {arquivo_entrada} lido com sucesso!")
        
        # Adiciona coluna com o nome do arquivo
        df['Origem'] = arquivo_entrada
        
        # Verificação da leitura e contagem de frequência
        try:
            frequencia_servico = df['DESCRIÇÃO DE SERVIÇO'].value_counts()
            print("Frequência dos serviços:\n", frequencia_servico)
        except KeyError as e:
            print(f"Erro: a coluna {e} não foi encontrada no DataFrame.")
        
        # Filtragem das linhas com palavras-chave
        criterios = ['chave fusivel','chave faca','chave by-pass', 'para-raios', 'poste', 'ramal', 'cabo', 'religador']
        try:
            mascara1 = df['DESCRIÇÃO DE SERVIÇO'].str.contains('|'.join(criterios), case=False, na=False)
            df_filtrado = df[mascara1]
            print(f"Arquivo filtrado {arquivo_entrada}:")
            display(df_filtrado)
        except KeyError as e:
            print(f"Erro: a coluna {e} não foi encontrada no DataFrame.")
        
        # Contagem de frequência dos serviços filtrados
        if not df_filtrado.empty:
            frequencia_servico_filtrado = df_filtrado['DESCRIÇÃO DE SERVIÇO'].value_counts()
            print("Frequência dos serviços filtrados:\n", frequencia_servico_filtrado)
        else:
            print("Nenhum dado filtrado encontrado.")
        
        # Exibição dos nomes das colunas após a filtragem
        if not df_filtrado.empty:
            print("\nNomes das colunas após a filtragem:")
            print(df_filtrado.columns)
        
        # Exclusão de colunas desnecessárias
        excluir_colunas = ['PEP&ORDEM&SERVIÇO','PI','EQUIPE','TIPO','TIPO DE MEDIÇÃO','Tipo de Medição','PEP IV NÍVEL','STATUS PEP', 'UN.','PU', 'OBS EQTL','Unnamed: 44', 'TIPO DE MEDIÇÃO.1',
                          'STATUS PEP\n(ALTERADO)','ORDEM/DIAGRAMA','STATUS ORDEM/DIAGRAMA','NOTA PROJ', 'Unnamed: 41', 'Unnamed: 45', 'TIPO DE MEDIÇÃO.1',
                          'ÁREA                                     DISTRIBUIDORA','CADERNO','ITEM','CÓD. SERVIÇO','CÓD. SERVIÇO&ITEM', 'TIPO DE MEDIÇÃO.2',
                          'QUANT / KM','QUANT / KM.1','TIPO DE MEDIÇÃO','PERÍODO DE MEDIÇÃO','TÉCNICO OBRA','NDS ', 'Unnamed: 42', 'Área Resp (DISTRIBUIDORA)',
                          'SUPERVISOR/SOLICITANTE','MUNICÍPIO','MUNICIPIO','DESLOCAMENTO\nDE','DESLOCAMENTO\nPARA', 'Unnamed: 43', 'Unnamed: 23', 'CONTROLE\n(DADOS ENGENHARIA)'
                          'EQUIPE','ESTRUTURA','CONTROLE\n(dados Engenharia)','NOTA\nRECLAMAÇÃO\n(NR)', 'NOTA\nRECLAMAÇÃO / OCORRÊNCIA\n(NR / OC)',
                          'PEDIDO','OBS PARCEIRA','REGIONAL','MEDIÇÃO','OBRA ','OBRA','TÉCNICO RESPONSÁVEL','Área Resp\n(DISTRIBUIDORA)','TIPO DE MEDIÇÃO.3']
        try:
            df_final = df_filtrado.drop(columns=excluir_colunas, errors='ignore')
            print("Colunas excluídas com sucesso!")
        except Exception as e:
            print(f"Erro ao excluir colunas: {e}")
        
        return df_final
    except Exception as e:
        print(f"Erro ao processar o arquivo {arquivo_entrada}: {e}")
        return pd.DataFrame()
```

```python
# Lista para armazenar DataFrames filtrados
todos_dados_filtrados = []

for arquivo in arquivos_entrada:
    df_processado = processar_arquivo(arquivo)
    if not df_processado.empty:
        todos_dados_filtrados.append(df_processado)

# Unificação dos DataFrames filtrados
if todos_dados_filtrados:
    df_unificado = pd.concat(todos_dados_filtrados, ignore_index=True)
    print("Dados unificados com sucesso!")
else:
    df_unificado = pd.DataFrame()
```

```python
# Salvando o resultado em uma nova planilha
novo_output_file = r'C:\Users\E1180062\Downloads\Análise de Orçamento\Dados Unificados Regionais\Dados_Unificados_FLO.xlsx'
try:
    df_unificado.to_excel(novo_output_file, index=False, sheet_name="SUL")
    print(f"\nArquivo salvo com sucesso em: {novo_output_file}")
except Exception as e:
    print(f"Erro ao salvar o arquivo: {e}")
```
