```python
import xlwings as xw
import pandas as pd
```

```python
# Parte 1 - Leitura do arquivo Excel
output_file = r'C:\Users\E1180062\Downloads\Análise de Orçamento\Dados_Combinados_ABR_PHB.xlsx'
try:
    arquivo_analise = pd.read_excel(output_file)
    print("Arquivo lido com sucesso!")
except Exception as e:
    print(f"Erro ao ler o arquivo: {e}")
```

```python
# Parte 2 - Verificação da leitura e contagem de frequência (opcional)DESCRIÇÃO DE SERVIÇO
if 'arquivo_analise' in locals():
    try:
        frequencia_servico = arquivo_analise['DESCRIÇÃO DE SERVIÇO'].value_counts()
        print("Frequência dos serviços:\n", frequencia_servico)
    except KeyError as e:
        print(f"Erro: a coluna {e} não foi encontrada no DataFrame.")
else:
    print("A variável 'arquivo_analise' não foi definida.")
```

```python
# Parte 3 - Filtragem das linhas com palavras-chave
criterios = ['chave fusivel','chave faca','chave by-pass', 'para-raios', 'poste', 'ramal', 'cabo', 'religador']
try:
    mascara = arquivo_analise['DESCRIÇÃO DE SERVIÇO'].str.contains('|'.join(criterios), case=False, na=False)
    arquivo_filtrado = arquivo_analise[mascara]
    print("Arquivo filtrado:")
    display(arquivo_filtrado)
except KeyError as e:
    print(f"Erro: a coluna {e} não foi encontrada no DataFrame.")
```

```python
# Parte 4 - Contagem de frequência dos serviços filtrados
if 'arquivo_filtrado' in locals():
    frequencia_servico_filtrado = arquivo_filtrado['DESCRIÇÃO DE SERVIÇO'].value_counts()
    print("Frequência dos serviços filtrados:\n", frequencia_servico_filtrado)
else:
    print("A variável 'arquivo_filtrado' não foi definida.")
```

```python
# Parte 5 - Exibição dos nomes das colunas após a filtragem
if 'arquivo_filtrado' in locals():
    print("\nNomes das colunas após a filtragem:")
    print(arquivo_filtrado.columns)
else:
    print("A variável 'arquivo_filtrado' não foi definida.")
```

```python
# Parte 6 - Exclusão de colunas desnecessárias
excluir_coluna_arquivo = ['PEP&ORDEM&SERVIÇO','PI','TIPO','Tipo de Medição','PEP IV NÍVEL','STATUS PEP', 'UN.','PU', 'OBS EQTL','Unnamed: 44',
                          'STATUS PEP\n(ALTERADO)','ORDEM/DIAGRAMA','STATUS ORDEM/DIAGRAMA','NOTA PROJ', 'Unnamed: 41', 'Unnamed: 45',
                          'ÁREA                                     DISTRIBUIDORA','CADERNO','ITEM','CÓD. SERVIÇO','CÓD. SERVIÇO&ITEM',
                          'QUANT / KM','QUANT / KM.1','TIPO DE MEDIÇÃO','PERÍODO DE MEDIÇÃO','TÉCNICO OBRA','NDS ', 'Unnamed: 42', 'Área Resp (DISTRIBUIDORA)',
                          'SUPERVISOR/SOLICITANTE','MUNICÍPIO','MUNICIPIO','DESLOCAMENTO\nDE','DESLOCAMENTO\nPARA', 'Unnamed: 43', 'Unnamed: 23',
                          'EQUIPE','ESTRUTURA','CONTROLE\n(dados Engenharia)','NOTA\nRECLAMAÇÃO\n(NR)', 'NOTA\nRECLAMAÇÃO / OCORRÊNCIA\n(NR / OC)',
                          'PEDIDO','OBS PARCEIRA','REGIONAL','MEDIÇÃO','OBRA ','OBRA','TÉCNICO RESPONSÁVEL','Área Resp\n(DISTRIBUIDORA)']
try:
    arquivo_final = arquivo_filtrado.drop(columns=excluir_coluna_arquivo, errors='ignore')
    print("Colunas excluídas com sucesso!")
except Exception as e:
    print(f"Erro ao excluir colunas: {e}")
```

```python
# Parte 7 - Exibição dos nomes das colunas após a exclusão
if 'arquivo_final' in locals():
    print("\nNomes das colunas após a filtragem:")
    print(arquivo_final.columns)
else:
    print("A variável 'arquivo_final' não foi definida.")
```

```python
# Parte 8 - Salvando o resultado em uma nova planilha
novo_output_file = r'C:\Users\E1180062\Downloads\Análise de Orçamento\Dados_Filtrados.xlsx'
try:
    arquivo_final.to_excel(novo_output_file, index=False, sheet_name="Serviços Filtrados") # type: ignore
    print(f"\nArquivo salvo com sucesso em: {novo_output_file}")
except Exception as e:
    print(f"Erro ao salvar o arquivo: {e}")    
```
